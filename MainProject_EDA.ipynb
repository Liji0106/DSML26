# DATA PREPROCESSING

```python
random_heights = np.random.randint(150, 181, size=len(data_employee))
data_employee['Height'] = random_heights
data_employee['Height']
```

# ANALYSING

```python
data_employee['Team'].value_counts() #Employees distribution across each team
```

```python
#Total number of employee

total= len(data_employee)
print("The total number of employee is : " ,total)
```

```python
#percentage split relative to the total no. of employees 

percentage_split = ((data_employee['Team'].value_counts())/total)*100
print("Percentage Split : ",percentage_split)
```

```python
#no. of employees based on their positions

employee_position=data_employee['Position'].value_counts()
employee_position
```

```python
#grouping employees based on age
bins = [20, 30, 40, 50, 60]
labels = ['20-29', '30-39', '40-49', '50-59']
data_employee['AgeGroup'] = pd.cut(data_employee['Age'], bins=bins, labels=labels, right=False)
age_group_distribution = data_employee['AgeGroup'].value_counts()
print("Age Group Distribution: " ,age_group_distribution)
```

```python
#Salary expenditure of each team
salary_team=data_employee.groupby('Team')['Salary'].sum()
print("Salary Expenditure of each Team is : ",salary_team)
   
```

```python
 # Team having highest salary expenditure
highest_team_salary=data_employee.groupby('Team')['Salary'].sum().idxmax() 
print("The Team having Highest Salary Expenditure : ",highest_team_salary)
```

```python
#Salary expenditure of each position
salary_position=data_employee.groupby('Position')['Salary'].sum()
print("Salary Expenditure of each Position : ",salary_position)
   
```

```python
# # position having highest salary expenditure
highest_position_salary=data_employee.groupby('Position')['Salary'].sum().idxmax() 
print("The Position having Highest Salary Expenditure: ",highest_position_salary)
```

```python

correlation = data_employee['Age'].corr(data_employee['Salary'])
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Salary', data=data_employee)
plt.title('Correlation between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
print('Correlation between age and salary:', correlation)
```

# GRAPHICAL REPRESENTATION

```python
distribution = data_employee['Team'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(distribution, labels=distribution.index)
plt.title('Distribution of Employees In Each Teams')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Position', data=data_employee)
plt.title('Employees Based on Positions')
plt.xlabel('Position')
plt.ylabel('No. of Employees')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
labels = age_group_distribution.index
values = age_group_distribution.values
bars = plt.bar(labels,values,color='blue',width=0.5)
plt.title('Age Group Distribution of Employees')
plt.xlabel('Age Group')
plt.ylabel('No. of Employees')
plt.show()
```

```python
team_expenditure = data_employee.groupby('Team')['Salary'].sum() 
plt.figure(figsize=(8, 6))
labels = team_expenditure.index
values = team_expenditure.values
bars = plt.bar(labels,values,color='green',width=0.4)
plt.title('Salary Expenditure by Team')
plt.xlabel('Team')
plt.ylabel('Total Salary Expenditure')
plt.xticks(rotation=90)
plt.show()
```

```python
position_expenditure = data_employee.groupby('Position')['Salary'].sum() 
plt.figure(figsize=(8, 6))
labels = position_expenditure.index
values = position_expenditure.values
bars = plt.bar(labels,values,color='red',width=0.5)
plt.title('Salary Expenditure by Team')
plt.xlabel('Team')
plt.ylabel('Total Salary Expenditure')
plt.show()
```
